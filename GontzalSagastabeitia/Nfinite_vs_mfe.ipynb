```python
import sys
import mfg_SIRC_functions as py
```

```python
gamma=0.9; rho=0.5; SIR0=(0.5,0.5,0); T=10; ci=500



params=py.Parameters(float('inf'))
C=2000
params.gamma=gamma/C
params.rho=rho/C
params.SIR0 = SIR0
params.T=C*T
params.ci=ci
params.A=[params.epsilon,1/C]
mfe_cost,mfe=py.compute_MFE_cost(params)



v_N=[4,8,12,16,20,24,28,32,50,100]
v_costs=[0]*len(v_N)

for n in range(len(v_N)):
    params=py.Parameters(v_N[n])
    params.gamma=gamma
    params.rho=rho
    params.T=T
    params.ci=ci
    BR,Vs,Vi=py.compute_MFE_N_players(params,1000)
    costBR=py.BR_cost_N_players(params,BR,Vs,Vi)/params.C
    v_costs[n]=costBR

import matplotlib.pyplot as plt

fig=plt.figure()
fig.set_size_inches((8,5))

plt.plot(v_N,v_costs, label="N<Inf")
plt.plot(v_N,[mfe_cost/C]*len(v_N),'-.', label="N=Inf")

plt.legend(loc="lower right")
plt.title("Fixed point cost for N<Inf")

plt.draw()
```
