```python
import numpy as np
import statistics
import scipy.stats as st
def JSQ(bezero_kopurua,lambda_,mu,c): 
    #bezeroen iritsiera denborak
    ir_a_den=[]
    for i in range(bezero_kopurua):
        ir_a_den.append(
            np.random.exponential(1/lambda_))

    iritsi_den=[]
    for i in range(bezero_kopurua):
        if i==0:
            iritsi_den.append(ir_a_den[0])
        else:
            iritsi_den.append(
                iritsi_den[i-1]+ir_a_den[i])

    #zerbitzu denbora
    zer_den=[]
    for i in range(bezero_kopurua):
        den=np.random.exponential(1/mu)
        zer_den.append(den)
 
    soj=[]   
    irteerak=[]
    for i in range(c):
        soj.append([])
        irteerak.append([])
    
    bezero_indizea=0
    ilara_bezero=[0 for i in range(c)]
    ilara_bezero_ind=[0 for i in range(c)]
    irteera_denbora=[-1 for i in range(c)]
    itxaron_denbora=[]

    while bezero_indizea<bezero_kopurua:
        
        for i in range(c):
            lu=len(irteerak[i])
            for j in range(lu):
                if (irteerak[i][0]
                    <iritsi_den[bezero_indizea]):
                    ilara_bezero[i]-=1
                    irteerak[i].pop(0)
                else:
                    break
        
        bezero_min=min(ilara_bezero)
        ilarak=[]
        for i in range(c):
            if bezero_min==ilara_bezero[i]:
                ilarak.append(i)
        n=np.random.choice(ilarak)
        
        bezero_indizea_ilaran=ilara_bezero_ind[n]
        
        if iritsi_den[bezero_indizea]>irteera_denbora[n]:  
            itxaron_denbora.append(0)
            soj[n].append(zer_den[bezero_indizea])
            irteera_denbora[n]=(soj[n][bezero_indizea_ilaran]
                                +iritsi_den[bezero_indizea])
            
        
        else:   
            itxaron_denbora.append(irteera_denbora[n]
                                   -iritsi_den[bezero_indizea])
            soj[n].append(zer_den[bezero_indizea]
                          +itxaron_denbora[bezero_indizea])
            irteera_denbora[n]=(soj[n][bezero_indizea_ilaran]
                                +iritsi_den[bezero_indizea])
        
        irteerak[n].append(irteera_denbora[n])    
        bezero_indizea=bezero_indizea+1
        ilara_bezero[n]=ilara_bezero[n]+1
        ilara_bezero_ind[n]=ilara_bezero_ind[n]+1
        
        
    ES=0
    for i in range(c):
        ES=ES+statistics.mean(soj[i])
    ES=ES/c
    EL=lambda_*ES
    EW=ES-1/mu
    return [ES,EL,EW]
 
def JSQ_kon(bezero_kopurua,lambda_,mu,c):
    ES=[]
    EL=[]
    EW=[]
    for i in range(15):
        emaitza=JSQ(bezero_kopurua,lambda_,mu,c)
        ES.append(emaitza[0])
        EL.append(emaitza[1])
        EW.append(emaitza[2])
    ES_tartea=st.t.interval(alpha = 0.95, 
                            df = len (ES) -1, 
                            loc = np.mean (ES),
                            scale = st.sem (ES))
    EL_tartea=st.t.interval(alpha = 0.95, 
                            df = len (EL) -1, 
                            loc = np.mean (EL),
                            scale = st.sem (EL))
    EW_tartea= st.t.interval(alpha = 0.95,
                             df = len (EW) -1,
                             loc = np.mean (EW),
                             scale = st.sem (EW))
    print("Simulazioan lortutakoa")
    print("ES=",statistics.mean(ES))
    print("EL=",statistics.mean(EL))
    print("EW=",statistics.mean(EW))
    print("Konfiantza tarteak")
    print("ES:",ES_tartea)
    print("EL:",EL_tartea)
    print("EW:",EW_tartea)
        

JSQ_kon(10**7,8,1,10)
```
