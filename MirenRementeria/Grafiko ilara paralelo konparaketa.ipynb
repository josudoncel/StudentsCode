```python
import numpy as np
import matplotlib.pyplot as plt
import statistics
import scipy.stats as st
import random

def BPP(bezero_kopurua,iritsi_denbora,zer_denbora,c):
    
    soj=[]
    for i in range(c):
        soj.append([])

    bezero_indizea=0
    irteera_denbora=[-1 for i in range(c)]
    itxaron_denbora=[]
    ilara_bezero=[0 for i in range(c)]
    

    while bezero_indizea<bezero_kopurua:
        n=np.random.choice([i for i  in range(c)])
        bezero_indize_ilaran=ilara_bezero[n]

        if iritsi_denbora[bezero_indizea]>irteera_denbora[n]:    
            itxaron_denbora.append(0)
            soj[n].append(zer_denbora[bezero_indizea])
            irteera_denbora[n]=(soj[n][bezero_indize_ilaran]
                                +iritsi_denbora[bezero_indizea])
        
        else:   #ilara ez dago libre 
            itxaron_denbora.append(irteera_denbora[n]
                                   -iritsi_denbora[bezero_indizea])
            soj[n].append(zer_denbora[bezero_indizea]
                          +itxaron_denbora[bezero_indizea])
            irteera_denbora[n]=(soj[n][bezero_indize_ilaran]
                                +iritsi_denbora[bezero_indizea])
    
        bezero_indizea=bezero_indizea+1
        ilara_bezero[n]=ilara_bezero[n]+1  
        
    ES=0
    for i in range(c):
        if len(soj[i])!=0:
            ES=ES+statistics.mean(soj[i])
    ES=ES/c

    return ES
        
def PO2(bezero_kopurua,iritsi_den,zer_den,c):
    soj=[]   
    irteerak=[]
    for i in range(c):
        soj.append([])
        irteerak.append([])
    
    bezero_indizea=0
    ilara_bezero=[0 for i in range(c)]
    ilara_bezero_ind=[0 for i in range(c)]
    irteera_denbora=[-1 for i in range(c)]
    itxaron_denbora=[]
    zerbidore=[i for i in range(c)]

    while bezero_indizea<bezero_kopurua:

        aukeratuak=random.choices(zerbidore, k=2)
        aukeratuak_bezero=[]
        for i in aukeratuak:
            lu=len(irteerak[i])
            for j in range(lu):
                if irteerak[i][0]<iritsi_den[bezero_indizea]:
                    ilara_bezero[i]-=1
                    irteerak[i].pop(0)
                else:
                    break
            aukeratuak_bezero.append(ilara_bezero[i])
        
        bezero_min=min(aukeratuak_bezero)
        ilarak=[]
        
        for i in range(2):
            if bezero_min==aukeratuak_bezero[i]:
                ilarak.append(aukeratuak[i])
    
        n=np.random.choice(ilarak) 
        bezero_indizea_ilaran=ilara_bezero_ind[n]
        
        if iritsi_den[bezero_indizea]>irteera_denbora[n]:  
            itxaron_denbora.append(0)
            soj[n].append(zer_den[bezero_indizea])
            irteera_denbora[n]=(soj[n][bezero_indizea_ilaran]
                                +iritsi_den[bezero_indizea])
            
        
        else:   #ilara ez dago libre 
            itxaron_denbora.append(irteera_denbora[n]
                                   -iritsi_den[bezero_indizea])
            soj[n].append(zer_den[bezero_indizea]
                          +itxaron_denbora[bezero_indizea])
            irteera_denbora[n]=(soj[n][bezero_indizea_ilaran]
                                +iritsi_den[bezero_indizea])
        
        irteerak[n].append(irteera_denbora[n])    
        bezero_indizea=bezero_indizea+1
        ilara_bezero[n]=ilara_bezero[n]+1
        ilara_bezero_ind[n]=ilara_bezero_ind[n]+1
        
    ES=0
    for i in range(c):
        if len(soj[i])!=0:
            ES=ES+statistics.mean(soj[i])
    ES=ES/c
    return ES
   
def JSQ(bezero_kopurua,iritsi_den,zer_den,c):
    soj=[]   
    irteerak=[]
    for i in range(c):
        soj.append([])
        irteerak.append([])
    
    bezero_indizea=0
    ilara_bezero=[0 for i in range(c)]
    ilara_bezero_ind=[0 for i in range(c)]
    irteera_denbora=[-1 for i in range(c)]
    itxaron_denbora=[]

    while bezero_indizea<bezero_kopurua:
        
        for i in range(c):
            lu=len(irteerak[i])
            for j in range(lu):
                if irteerak[i][0]<iritsi_den[bezero_indizea]:
                    ilara_bezero[i]-=1
                    irteerak[i].pop(0)
                else:
                    break
        
        bezero_min=min(ilara_bezero)
        ilarak=[]
        for i in range(c):
            if bezero_min==ilara_bezero[i]:
                ilarak.append(i)
        n=np.random.choice(ilarak)
        
        bezero_indizea_ilaran=ilara_bezero_ind[n]
        
        if iritsi_den[bezero_indizea]>irteera_denbora[n]:  #ilara libre dago  
            itxaron_denbora.append(0)
            soj[n].append(zer_den[bezero_indizea])
            irteera_denbora[n]=(soj[n][bezero_indizea_ilaran]
                                +iritsi_den[bezero_indizea])
            
        
        else:   #ilara ez dago libre 
            itxaron_denbora.append(irteera_denbora[n]
                                   -iritsi_den[bezero_indizea])
            soj[n].append(zer_den[bezero_indizea]
                          +itxaron_denbora[bezero_indizea])
            irteera_denbora[n]=(soj[n][bezero_indizea_ilaran]
                                +iritsi_den[bezero_indizea])
        
        irteerak[n].append(irteera_denbora[n])    
        bezero_indizea=bezero_indizea+1
        ilara_bezero[n]=ilara_bezero[n]+1
        ilara_bezero_ind[n]=ilara_bezero_ind[n]+1
        
        
    ES=0
    for i in range(c):
        if len(soj[i])!=0:
            ES=ES+statistics.mean(soj[i])
    ES=ES/c
    return ES

def grafika(lambda_,mu):
    bezero=[]
    BPP_=[]
    PO2_=[]
    JSQ_=[]
    #bezeroen iritsiera denborak
    ir_a_den=[]
    for j in range(10**6):
        ir_a_den.append(
            np.random.exponential(1/lambda_))

    iritsi_den=[]
    for j in range(10**6):
        if j==0:
            iritsi_den.append(ir_a_den[0])
        else:
            iritsi_den.append(
                iritsi_den[j-1]+ir_a_den[j])

        #zerbitzu denbora
    zer_den=[]
    for j in range(10**6):
        den=np.random.exponential(1/mu)
        zer_den.append(den)      
        
    for i in range(10,10**6,1000):
        bezero.append(i)
        BPP_.append(BPP(i,iritsi_den,zer_den,10))
        PO2_.append(PO2(i,iritsi_den,zer_den,10))
        JSQ_.append(JSQ(i,iritsi_den,zer_den,10))
        
    plt.plot(bezero,BPP_,"b",bezero,PO2_,"orange",bezero,JSQ_,"g")
    plt.xlabel('Bezero kopurua')
    plt.ylabel('E[S]')
    plt.legend(("BP",'Po2','JSQ'),loc="lower right")
    plt.savefig('c10bezero.pdf')

lambda_=0.8*10
mu=1
grafika(lambda_,mu)
```
