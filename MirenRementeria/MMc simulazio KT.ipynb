```python
import numpy as np
import matplotlib.pyplot as plt
import statistics
from math import factorial
import scipy.stats as st


def MMc(bezero_kopurua,lambda_,mu,c):
    #bezeroen iritsiera denborak
    ir_a_den=[]
    for i in range(bezero_kopurua):
        ir_a_den.append(
            np.random.exponential(1/lambda_))
    iritsi_den=[]
    for i in range(bezero_kopurua):
        if i==0:
            iritsi_den.append(ir_a_den[i])
        else:
            iritsi_den.append(iritsi_den[i-1]
                              +ir_a_den[i])
    #zerbitzu denbora
    zer_den=[]
    for i in range(bezero_kopurua):
        den=np.random.exponential(1/mu)
        zer_den.append(den)
    bezero_indizea=0
    sojourn=0
    irteera_denbora=[-1 for i in range(c)]
    soj=[]
    itxaron_denbora=[]
    while bezero_indizea<bezero_kopurua:
        zerbitzuan_sartu=False 
        for i in range(c): #ikusi ilara dagoen 
            if (iritsi_den[bezero_indizea]
                >irteera_denbora[i]):
                zerbitzuan_sartu=True 
                itxaron_denbora.append(0)
                soj.append(zer_den[bezero_indizea])
                irteera_denbora[i]=(
                    iritsi_den[bezero_indizea]
                            +soj[bezero_indizea])
                break  
        if not zerbitzuan_sartu: 
            zerbitzaria=np.argmin(irteera_denbora)
            itxaron_denbora.append(
                irteera_denbora[zerbitzaria]
                            -iritsi_den[bezero_indizea])
            soj.append(
                zer_den[bezero_indizea]
                       +itxaron_denbora[bezero_indizea])
            irteera_denbora[zerbitzaria]=(
                iritsi_den[bezero_indizea]
                                 +soj[bezero_indizea])
        bezero_indizea+=1
    
    ES=statistics.mean(soj)
    EL=lambda_*ES
    EW=statistics.mean(itxaron_denbora)
    return [ES,EL,EW]    
def MMc_kon(bezero_kopurua,lambda_,mu,c):
    rho=lambda_/(c*mu)
    #Teoriako emaitzak 
    batuketa=0
    for i in range(c):
        batuketa=batuketa+((c*rho)**i)/factorial(i)
    d_p=((((c*rho)**c)/factorial(c))*((1-rho)*batuketa
                    + ((c*rho)**c)/factorial(c))**(-1))
    L_q=d_p*(rho/(1-rho))
    W=d_p/(c*mu)+L_q/(c*mu)
    S=W+1/mu
    L=lambda_*S
    print("Teorian")
    print("ES=",S)
    print("EL=",L)
    print("EW=",W)
    ES=[]
    EL=[]
    EW=[]
    for i in range(15):
        emaitza=MMc(bezero_kopurua,lambda_,mu,c)
        ES.append(emaitza[0])
        EL.append(emaitza[1])
        EW.append(emaitza[2])
    ES_tartea=st.t.interval(alpha = 0.95, 
                            df = len (ES) -1, 
                            loc = np.mean (ES),
                            scale = st.sem (ES))
    EL_tartea=st.t.interval(alpha = 0.95, 
                            df = len (EL) -1, 
                            loc = np.mean (EL),
                            scale = st.sem (EL))
    EW_tartea= st.t.interval(alpha = 0.95, 
                             df = len (EW) -1, 
                             loc = np.mean (EW),
                             scale = st.sem (EW))
    print("Simulazioan")
    print("ES=",statistics.mean(ES))
    print("EL=",statistics.mean(EL))
    print("EW=",statistics.mean(EW))
    print("Konfiantza tarteak")
    print("ES:",ES_tartea)
    print("EL:",EL_tartea)
    print("EW:",EW_tartea)    
MMc_kon(10**7,2,1.5,3)
```
