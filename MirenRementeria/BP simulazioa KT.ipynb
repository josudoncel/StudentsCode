```python
import numpy as np
import statistics
import scipy.stats as st
def BP(bezero_kopurua,lambda_,mu,N):
    #iritsi eta zerbitzu denborak
    iritsi_denbora=[]
    zer_denbora=[]
    iritsi=[]
    ilara=[]
    soj=[]
    for i in range(N):
        iritsi_denbora.append([])
        zer_denbora.append([])
        soj.append([])
    for i in range(bezero_kopurua):
        if i==0:
            iritsi.append(
                np.random.exponential(1/lambda_))
        else:
            iritsi.append(iritsi[i-1]
                    +np.random.exponential(1/lambda_))
            
        n=np.random.choice(
            [i for i in range(N)])
        iritsi_denbora[n].append(iritsi[i])
        zer_denbora[n].append(
            np.random.exponential(1/mu))
        ilara.append(n)
        

    bezero_indizea=0
    irteera_denbora=[-1 for i in range(N)]
    itxaron_denbora=[]
    ilara_bezero=[0 for i in range(N)]
    

    while bezero_indizea<bezero_kopurua:
        n=ilara[bezero_indizea]
        bezero_indize_ilaran=ilara_bezero[n]

        if iritsi_denbora[n][
            bezero_indize_ilaran]>irteera_denbora[n]: 
            itxaron_denbora.append(0)
            soj[n].append(
                zer_denbora[n][bezero_indize_ilaran])
            irteera_denbora[n]=(soj[n][
                bezero_indize_ilaran]+iritsi_denbora[n][
                bezero_indize_ilaran])
        
        else:   
            itxaron_denbora.append(
                irteera_denbora[n]-iritsi_denbora[n][
                    bezero_indize_ilaran])
            soj[n].append(
                zer_denbora[n][bezero_indize_ilaran]+
                itxaron_denbora[bezero_indizea])
            irteera_denbora[n]=(soj[n][bezero_indize_ilaran]
            +iritsi_denbora[n][bezero_indize_ilaran])
    
        bezero_indizea=bezero_indizea+1
        ilara_bezero[n]=ilara_bezero[n]+1  
        
    ES=0
    for i in range(N):
        ES=ES+statistics.mean(soj[i])
    ES=ES/N
    EL=lambda_*ES
    EW=ES-1/mu
    return [ES,EL,EW]

def BP_kon(bezero_kopurua,lambda_,mu,N):
    rho=lambda_/mu
    print("Teorian")
    ES=1/(mu-lambda_/N)
    print("ES=",ES)
    print("EL=",ES*lambda_)
    print("EW=",ES-1/(mu))
    ES=[]
    EL=[]
    EW=[]
    for i in range(15):
        emaitza=BP(bezero_kopurua,lambda_,mu,N)
        ES.append(emaitza[0])
        EL.append(emaitza[1])
        EW.append(emaitza[2])
    ES_tartea=st.t.interval(alpha = 0.95, 
                            df = len (ES) -1, 
                            loc = np.mean (ES),
                            scale = st.sem (ES))
    EL_tartea=st.t.interval(alpha = 0.95,
                            df = len (EL) -1, 
                            loc = np.mean (EL), 
                            scale = st.sem (EL))
    EW_tartea= st.t.interval(alpha = 0.95, 
                             df = len (EW) -1, 
                             loc = np.mean (EW), 
                             scale = st.sem (EW))
    print("Simulazioan lortutakoa")
    print("ES=",statistics.mean(ES))
    print("EL=",statistics.mean(EL))
    print("EW=",statistics.mean(EW))
    print("Konfiantza tarteak")
    print("ES:",ES_tartea)
    print("EL:",EL_tartea)
    print("EW:",EW_tartea)
        
   
BP_kon(10**7,8,1,10)  
```
