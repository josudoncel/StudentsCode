```python
import numpy as np
import statistics
import scipy.stats as st
def MM1(bezero_kopurua,lambda_,mu):
    #bezeroen iritsiera denborak
    ir_a_den=[]
    for i in range(bezero_kopurua):
        ir_a_den.append(
            np.random.exponential(1/lambda_))
    iritsi_den=[]
    for i in range(bezero_kopurua):
        if i==0:
            iritsi_den.append(ir_a_den[0])
        else:
            iritsi_den.append(
                iritsi_den[i-1]+ir_a_den[i])
    #zerbitzu denbora
    zer_den=[]
    for i in range(bezero_kopurua):
        den=np.random.exponential(1/mu)
        zer_den.append(den)
    bezero_indizea=0
    irteera_denbora=-1
    soj=[]
    itxaron_denbora=[]
    while bezero_indizea<bezero_kopurua:
        if iritsi_den[bezero_indizea]>irteera_denbora:  
            itxaron_denbora.append(0)
            soj.append(zer_den[bezero_indizea])
            irteera_denbora=(soj[bezero_indizea]
                             +iritsi_den[bezero_indizea])
            bezero_indizea=bezero_indizea+1
        else:   
            itxaron_denbora.append(irteera_denbora
                                  -iritsi_den[bezero_indizea])
            soj.append(zer_den[bezero_indizea]
                       +itxaron_denbora[bezero_indizea])
            irteera_denbora=(soj[bezero_indizea]
                             + iritsi_den[bezero_indizea])
            bezero_indizea=bezero_indizea+1 
    ES=statistics.mean(soj)
    EL=lambda_*ES
    EW=ES-1/mu
    return([ES,EL,EW]) 
def MM1_kon(bezero_kopurua,lambda_,mu):
    rho=lambda_/mu
    print("Teorian")
    print("ES=",(1/mu)/(1-rho))
    print("EL=",rho/(1-rho))
    print("EW=",(rho/mu)/(1-rho))
    ES=[]
    EL=[]
    EW=[]
    for i in range(15):
        emaitza=MM1(bezero_kopurua,lambda_,mu)
        ES.append(emaitza[0])
        EL.append(emaitza[1])
        EW.append(emaitza[2])
    ES_tartea=st.t.interval(alpha = 0.95, 
                            df = len (ES) -1, 
                            loc = np.mean (ES), 
                            scale = st.sem (ES))
    EL_tartea=st.t.interval(alpha = 0.95, 
                            df = len (EL) -1,
                            loc = np.mean (EL), 
                            scale = st.sem (EL))
    EW_tartea= st.t.interval(alpha = 0.95,
                             df = len (EW) -1, 
                             loc = np.mean (EW), 
                             scale = st.sem (EW))
    print("Simulazioan lortutakoa")
    print("ES:",ES_tartea)
    print("EL:",EL_tartea)
    print("EW:",EW_tartea)    
    
MM1_kon(10**7,1.5,2) 
```
