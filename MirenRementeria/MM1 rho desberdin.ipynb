```python
import numpy as np
import statistics
import scipy.stats as st



def MM1(bezero_kopurua,lambda_,mu):
    #bezeroen iritsiera denborak
    ir_a_den=[]
    for i in range(bezero_kopurua):
        ir_a_den.append(
            np.random.exponential(1/lambda_))

    iritsi_den=[]
    for i in range(bezero_kopurua):
        if i==0:
            iritsi_den.append(ir_a_den[0])
        else:
            iritsi_den.append(
                iritsi_den[i-1]+ir_a_den[i])

    #zerbitzu denbora
    zer_den=[]
    for i in range(bezero_kopurua):
        den=np.random.exponential(1/mu)
        zer_den.append(den)

    bezero_indizea=0
    irteera_denbora=-1
    soj=[]
    itxaron_denbora=[]

    while bezero_indizea<bezero_kopurua:
        if iritsi_den[bezero_indizea]>irteera_denbora:  
            itxaron_denbora.append(0)
            soj.append(zer_den[bezero_indizea])
            irteera_denbora=(soj[bezero_indizea]
                             +iritsi_den[bezero_indizea])
            bezero_indizea=bezero_indizea+1
        
        else:   
            itxaron_denbora.append(irteera_denbora
                                  -iritsi_den[bezero_indizea])
            soj.append(zer_den[bezero_indizea]
                       +itxaron_denbora[bezero_indizea])
            irteera_denbora=(soj[bezero_indizea]
                             + iritsi_den[bezero_indizea])
            bezero_indizea=bezero_indizea+1

  
    ES=statistics.mean(soj)
    return ES
  
    
def MM1_kon(bezero_kopurua,lambda_,mu):
    rho=lambda_/mu
    print("Teorian")
    print("ES=",(1/mu)/(1-rho))
    ES=[]
    for i in range(15):
        emaitza=MM1(bezero_kopurua,lambda_,mu)
        ES.append(emaitza)
    ES_tartea=st.t.interval(alpha = 0.95, 
                            df = len (ES) -1, 
                            loc = np.mean (ES), 
                            scale = st.sem (ES))
    print("Simulazioan lortutakoa")
    print("ES=",statistics.mean(ES))
    print("Konfiantza tartea")
    print("ES:",ES_tartea)

for i in [0.1,0.25,0.5,0.75,0.9]:
    print("Rho=",i)
    MM1_kon(10**7,i,1)
    
```
